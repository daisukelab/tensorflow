```python
# http://www.slideshare.net/hoxo_m/ss-45361675
# この例題を自習する

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def Beta_average(a, b):
    return a / (float(a) + b)

def Beta_variance(a, b):
    return (a * b) / (((float(a) + b) ** 2) * (a + b + 1.0))

def Beta(p, a, b):
    def _base_value(p, a, b):
        return (p ** (a - 1)) * ((1 - p) ** (b - 1))
    ave_p = Beta_average(a, b)
    base_max = _base_value(ave_p, a, b)
    var = Beta_variance(a, b)
    # 規格化定数Cは、平均と分散から擬似計算する
    C = (1/base_max) / np.sqrt(var)
    print ave_p, var, C, base_max
    return C * _base_value(p, a, b)

t = np.arange(0.0, 1.0, 0.01)
plt.plot(t, Beta(t, 2, 2))
plt.ylabel('density')
plt.grid(True)
plt.show()
```

```python
# ベルヌーイ分布とBeta分布を組み合わせてプロットしてみる。見てみたいだけ
def L(p, x):
    return (p ** x) * ((1 - p) ** (1 - x))

plt.plot(t, Beta(t, 2, 2))
plt.plot(t, L(t, 1) * Beta(t, 2, 2))
plt.plot(t, L(t, 1))
plt.plot(t, L(t, 0))
plt.show()
```

```python
# 単に正規分布をプロット、見てみたいだけ
def NormDist(x, ave, std):
    return 1.0 / (2.0 * std ** 2 * np.pi) * np.exp(-(x - ave)**2 / (2.0 * std**2))
plt.plot(t, NormDist(t, 0.5, 0.1))
plt.show()
```

```python
# alphaを増やしてプロットしてみる。a,b = 1,1のとき --> 直線ですか…
a = 1.0
b = 1.0
for i in range(5):
    plt.plot(t, Beta(t, a, b))
    a += 1.0
plt.show()
```
